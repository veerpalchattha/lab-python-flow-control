```python
# 1. List Products
products = ["t-shirt", "mug", "hat", "book", "keychain"]
print(products)
inventory = {}
values = []

for i in range(0,5):
    user_num = input(f"Enter number {i+1}: ")
    if user_num.isdigit():
        values.append(user_num)
    else:
        print("Enter the valid value")
        user_num = input(f"Enter number {i+1}: ")
        values.append(user_num)
print(values)

for i in range(0,5):
    inventory[products[i]] = int(values[i])
print(inventory)
```

```python
customer_orders = set()

for i in range(0,3):
    # a. Prompt the user to enter the name of a product.
    product = input(f"Enter the product {i+1}: ")
    # b. Add the product to the "customer_orders" set.
    if product.isdigit():
        product = print(f"Enter the valid product {i+1}:")
        product = input(f"Enter the product {i+1} that you want to order : ")
        customer_orders.add(product)
    else:
        customer_orders.add(product)
    another_product = input("Do you want to add another product? (yes/no): ").strip().lower()

    # c. Continue only if yes
    if another_product != "yes":
        break


print("Customer orders:", customer_orders)
```

```python
products = []

for product in customer_orders:
    while True:
        quantity = input(f"Enter the quantity for the product {product}: ")
        if quantity.isdigit():
            quantity = int(quantity)
            break
        print("Please enter a valid numeric quantity.")
   
    products.append(quantity)

    # Update inventory if product exists
    if product in inventory:
        inventory[product] = int(inventory[product]) - quantity
    else:
        print(f"Warning: {product} not found in inventory!")

print(inventory)
```
